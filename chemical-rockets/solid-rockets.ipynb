# Solid Rocket Motors

Solid rocket motors store propellant as a solid "grain" containing both fuel and oxidizer. The grain burns from exposed surfaces, generating hot gas that is expelled through a nozzle.

## Motor Configuration

```{figure} #fig-solid-motor-schematic
:name: fig-solid-motor-schematic
:align: center

Cross-section of a solid rocket motor showing: motor casing (pressure vessel), propellant grain with internal cavity (port), web thickness (distance from burning surface to casing), igniter at forward end, gas cavity, and nozzle. Section A-A shows a star-shaped grain cross-section with burning surface $A_b$.
```

The propellant grain contains both **fuel and oxidizer** in a solid matrix.

Advantages of solid motors:
1. Simplicity — no pumps, valves, or plumbing
2. Compact — high propellant mass fraction
3. Storable — ready to fire on short notice (can be stored for years)

Disadvantages:

1. Lower $I_{sp}$ than liquid engines (typically 250–280 s)
2. No flexibility — cannot control $\dot{m}$, cannot stop/restart
3. Mass flow rate depends on chamber conditions: $\dot{m} = f(P_c, T_i)$
4. Very sensitive to grain defects (cracks, voids, bubbles) which cause anomalous burning

## Propellant Burning Rate

The **burning rate** $\dot{r}$ is the **surface regression velocity**---how fast the grain surface recedes normal to itself.

$$
\dot{r} \quad \left[\frac{\text{mm}}{\text{s}}\right]
$$

```{figure} #fig-burning-surface
:name: fig-burning-surface
:align: center

Diagram showing propellant grain with burning surface receding at rate $\dot{r}$, chamber pressure $P_c$, and total instantaneous burning area $A_b$.
```

The mass generation rate from propellant burning is:

$$
\boxed{\dot{m} = \dot{r} \, A_b \, \rho_p}
$$

where:
- $\dot{r}$ = burning rate (surface regression velocity)
- $A_b$ = instantaneous burning surface area
- $\rho_p$ = propellant density

## Grain Geometry and Thrust Profiles

Since thrust is:

$$
F = \dot{m} \, C_F \, c^*
$$

and $C_F$, $c^*$ are essentially **constant** for a given propellant and nozzle, the thrust profile depends on how $\dot{m}$ varies with time---which depends on $A_b(t)$.

```{figure} #fig-grain-geometries
:name: fig-grain-geometries
:align: center

Common grain geometries and their thrust profiles showing cross-sections and corresponding $F$ or $P_c$ versus time curves.
```

### Grain Types and Thrust Characteristics

| Geometry | Cross-Section | Burn Pattern | Thrust Profile | $A_b(t)$ |
|----------|---------------|--------------|----------------|----------|
| **Tubular** | Circular bore | Burns radially outward | Progressive (increasing) | Increases |
| **Bone & Tube** | Tube with slot | Combined surfaces | Neutral (constant) | ~Constant |
| **Star** | Multi-pointed star | Complex internal | Approximately neutral | ~Constant |
| **Double Anchor** | Anchor shape | Multiple surfaces | Regressive (decreasing) | Decreases |

:::{note}
A **six-pointed star** grain is very common for large boosters (e.g., Space Shuttle SRBs) because it provides high initial thrust with a nearly neutral profile.
:::

## Burning Rate Law

The burning rate is an **empirical function** of chamber pressure:

```{figure} #fig-burning-rate-physics
:name: fig-burning-rate-physics
:align: center

Burning rate physics showing: boundary layer above grain surface, flame zone with heat release, and heat transfer back to the solid propellant causing surface regression at rate $\dot{r}$. Chamber pressure $P_c$ affects flame properties and heat transfer.
```

An empirical relationship between burning rate and chamber pressure is:

$$
\boxed{\dot{r} = a \, P_c^n}
$$

where the pressure exponent $n$ is:
- Not a function of $P_c$ or $T_i$
- Measured experimentally for specific propellant combinations
- Typically $n < 1.0$ (usually 0.3–0.7)

and the temperature coefficient $a$:
- Strong function of initial temperature $T_i$: $\quad a = a(T_i)$
- Not a function of pressure
- Higher $T_i$ → higher $a$ → faster burning

```{figure} #fig-burning-rate-curves
:name: fig-burning-rate-curves
:align: center

Burning rate $\dot{r}$ versus chamber pressure $P_c$ on log-log axes for different initial grain temperatures. Higher initial temperature $T_i$ shifts the curve upward, giving higher burning rates at the same pressure. The slope of each line is the pressure exponent $n$.
```

:::{important}
The temperature sensitivity of the burning rate is critical for solid motor performance prediction. A motor loaded in cold weather will have different thrust characteristics than one loaded in hot weather.
:::

## Equilibrium Chamber Pressure

```{figure} #fig-solid-motor-balance
:name: fig-solid-motor-balance
:align: center

Mass balance in a solid rocket motor: propellant mass is generated by grain burning ($\dot{m}_p = \dot{r} A_b \rho_p$) and exits through the nozzle ($\dot{m}_{out} = P_c A_t / c^*$). The motor casing acts as a pressure vessel containing the combustion gases.
```

Performing a mass balance:

$$
\frac{dm}{dt} = \dot{m}_p - \dot{m}_{out}
$$

where:
- $\dot{m}_p = \dot{r} A_b \rho_p = a P_c^n A_b \rho_p$ (mass generation rate from burning)
- $\dot{m}_{out} = \dfrac{P_c A_t}{c^*}$ (mass flow through nozzle)


At equilibrium (quasi-steady-state operation): $\dfrac{dm}{dt} \approx 0$:

$$
\dot{r} A_b \rho_p - \frac{P_c A_t}{c^*} = 0
$$

Substituting the burning rate law:

$$
a P_c^n A_b \rho_p - \frac{P_c A_t}{c^*} = 0
$$

Solving for $P_c$:

$$
a P_c^n A_b \rho_p = \frac{P_c A_t}{c^*}
$$

$$
P_c^{n-1} = \frac{A_t}{c^* \rho_p a A_b}
$$

$$
\boxed{P_c = \left(\rho_p \, c^* \, a \, \frac{A_b}{A_t}\right)^{\frac{1}{1-n}}}
$$

This is the **equilibrium chamber pressure**.

## Klemmung (K-factor)

The ratio of burning area to throat area is called **Klemmung** ($K$) factor:

$$
K = \frac{A_b}{A_t} \gg 1
$$

Typical values: $K \approx 100$

This large ratio means a small throat area can control the chamber pressure generated by a much larger burning surface.

## Pressure-Time Behavior

```{figure} #fig-pressure-equilibrium
:name: fig-pressure-equilibrium
:align: center

Chamber pressure $P_c$ versus time showing the quasi-steady equilibrium. After ignition, pressure rises rapidly to equilibrium value, remains approximately constant during the quasi-steady burn phase (for neutral grain), then drops at burnout.
```

:::{important}
**Stability requirement:** Since $n < 1$, the equilibrium is **stable**.

- If $P_c$ increases slightly → $\dot{m}_{out}$ increases faster than $\dot{m}_p$ (because $\dot{m}_{out} \propto P_c$ but $\dot{m}_p \propto P_c^n$ with $n < 1$) → pressure returns to equilibrium

- If $n > 1$, the motor would be **unstable** (runaway pressure increase leading to catastrophic failure)

This is why propellant formulations are carefully designed to ensure $n < 1$.
:::
